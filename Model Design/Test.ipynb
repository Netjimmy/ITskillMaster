```python
import numpy as np
import numpy.linalg as LA
def getDocDistance(a, b):
    if LA.norm(a)==0 or LA.norm(b)==0:
        return -1
    return round(np.inner(a, b) / (LA.norm(a) * LA.norm(b)), 4)
```
